## **Ejercicio 2**
**Seleccionar y justificar las 4 variables más representativas.**


### Con correlación

Para seleccionar 4 variables representativas, utilizaremos la correlación; así, podremos encontrar las variables que más se relacionan con las otras. Las variables con alta correlación pueden ser indicativas de que están representando características importantes del conjunto de datos.

Para obtener esa correlación, eliminaremos completamente la columna "c" (columna con datos no numéricos) y obtendremos la martriz de correlación de Pearson.

```python
df2 = df.drop(columns=['c'])
correlacion = df2.corr()
correlacion
```

Seleccionamos las 4 variables que más se correlacionan entre sí:
1. Se convierte con valor absoluto los valores de su correlación.
2. Se suman los valores de cada variable (columnas). Los que tengan mayor valor, tienen una correlación mayor con las demás.
3. Se ordenan las variables de la mayor a la menor suma.
4. Se toman sólo las 4 más grandes.

```python
correlacion_fuerte = correlacion.abs().sum().sort_values(ascending=False).head(4)
print("Variables con mayor correlación:\n", correlacion_fuerte)
```

Encontramos entonces que "E3", "E5", "N10" y "E7" son las varibles con más correlación; por lo tanto, las que tomaremos.

Estas son importantes porque muestran una relación fuerte con otras variables, lo que indica que están capturando comportamientos similares o cruciales en el conjunto de datos.

### Con varianza mayor

Podemos hacer el mismo análisis con su varianza pues las variables con mayor varianza suelen contener más información, ya que muestran más diversidad en sus valores pues éstas contienen más información ya que no tienen valores constantes o cercanos entre sí, lo que las hace más representativas.

```python
varianza = df2.var().sort_values(ascending=False).head(4)
print("\nVariables con mayor varianza: \n", varianza)
```

Vemos entonces que las variables con más información serían: "a", "r", "s" y "E7".

Sin embargo, al corroborar los datos de la tabla, podemos ver cómo existen números muy atípicos. Esto lo podemos ver con un análisis con *box-plot* que nos muestra la basta diferencia entre los *ouliers* (valores atípicos) y el promedio:

```python
plt.figure(figsize=(4,3))
sns.boxplot(data=df[['s',"E3"]])
plt.title('Distribución de las variables seleccionadas')
plt.show()
```

```python
# Determinar cuántos valores faltantes existen por cada atributo
valores_faltantes_E3 = df[['E3']].isnull().sum()
print("Valores faltantes por columna E3")
print(valores_faltantes_E3)

# Determinar cuántos valores faltantes existen por cada atributo
valores_faltantes_E5 = df[['E5']].isnull().sum()
print("Valores faltantes por columna E5")
print(valores_faltantes_E5)

# Determinar cuántos valores faltantes existen por cada atributo
valores_faltantes_N10 = df[['N10']].isnull().sum()
print("Valores faltantes por columna N10")
print(valores_faltantes_N10)

# Determinar cuántos valores faltantes existen por cada atributo
valores_faltantes_E7 = df[['E7']].isnull().sum()
print("Valores faltantes por columna E7")
print(valores_faltantes_E7)
```

##### **Bonus** (no lo pide el ejercicio, pero me pareció interesante buscar cómo contar los *outliers*)

Si nos referimos a la cantidad de *outliers*:

```python
def contar_outliers(columna):
    Q1 = columna.quantile(0.25)
    Q3 = columna.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = columna[(columna < lower_bound) | (columna > upper_bound)]
    return len(outliers)

# Contar los outliers en cada una de las variables seleccionadas
outliers_E3 = contar_outliers(df['E3'])
outliers_E5 = contar_outliers(df['E5'])
outliers_N10 = contar_outliers(df['N10'])
outliers_E7 = contar_outliers(df['E7'])

print("\nNúmero de outliers por columna:")
print(f"E3: {outliers_E3}")
print(f"E5: {outliers_E5}")
print(f"N10: {outliers_N10}")
print(f"E7: {outliers_E7}")
```

## **Actividad 5**

**Ejemplificar un proceso de transformación para valores faltantes o datos
inconsistentes (“raros”).**


Son dos estrategías para intentar resolver los problemas "raros", podemos catalogarlos como **Datos Faltantes** y **Ouliers**.

Se puede practicamente de la misma manera para cada grupo: "Eliminar los datos o registros faltantes" o "Reemplazar o ingresar los valores no adecuados".

### Datos Faltantes

* **Eliminar los registros con valores faltantes:** Si hay pocos valores faltantes y no afectan mucho a la calidad del análisis.
* **Ingresar los valores:** Rellenar los valores faltantes con una estrategia, como la media, mediana, moda, o usando algún método más avanzado (como regresión).


### *Outliers*

* **Eliminar los outliers:** Si los valores extremos no tienen sentido o distorsionan los resultados.
* **Remplazar los outliers:** Puedes reemplazarlos por valores más razonables, como el límite superior o inferior del rango intercuartílico (IQR).
