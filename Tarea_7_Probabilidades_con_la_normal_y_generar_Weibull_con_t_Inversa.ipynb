### Cálculo de probabilidades:
$\mathbb{P}(0<Z<b)$: Queremos calcular $\mathbb{P}(0<Z<0.43)$, lo cuál puede realizarse de la siguiente manera:
* Tablas de área a la derecha: $\mathbb{P}(0<Z<0.43) = 0.1664$

$\mathbb{P}(-b<Z<b)$: Queremos calcular $\mathbb{P}(-0.16<Z<0.16)$, lo cuál puede realizarse de la siguiente manera:
* Tablas de área a la derecha:  $\mathbb{P}(-0.16<Z<0.16) = \mathbb{P}(-0.16<Z<0)+\mathbb{P} (0<Z<0.16) = \mathbb{P}(0<Z<0.16)+\mathbb{P}(0<Z<0.16) = 2\mathbb{P}(0<Z<0.16) = 2(0.0636) = 0.1272$

$\mathbb{P}(Z<-b)$: Queremos calcular $\mathbb{P}(Z<-1.94)$, lo cuál puede realizarse de la siguiente manera:
* Tablas de área a la derecha: $\mathbb{P}(Z<-1.94) = \mathbb{P}(Z<0)+\mathbb{P}(-1.94<Z<0) = \mathbb{P}(Z<0) + \mathbb{P}(0<Z<1.94) = 0.5 - 0.4738 = 0.0262$

$\mathbb{P}(Z>-b)$: Queremos calcular $(Z>-0.07)$, lo cuál puede realizarse de la siguiente manera:
* Tablas de área a la derecha: $\mathbb{P}(Z>-0.07) = \mathbb{P}(-0.07<Z<0) + \mathbb{P}(Z>0) = \mathbb{P}(0<Z<0.07) + \mathbb{P}(Z>0) = 0.0279 + 0.5 = 0.5279$

## Ejercicio 1
$\mathbb{P}(0<Z<b)$: Queremos calcular $\mathbb{P}(0<Z<0.43)$, lo cuál puede realizarse de la siguiente manera:

```python
norm(0,1).cdf(0.43)-norm(0,1).cdf(0)
```

```python
fig, ax = plt.subplots()
# for distribution curve
x= np.arange(-3,3,0.001)
ax.plot(x, norm.pdf(x,loc=0,scale=1))
ax.set_title("N(0,$1^2$)")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)
px=np.arange(0,0.43,0.01)
ax.set_ylim(0,0.45)
ax.fill_between(px,norm.pdf(px,loc=0,scale=1),alpha=0.5, color='g')
pro=norm(0,1).cdf(0.43)-norm(0,1).cdf(0)
ax.text(0.2,0.02,round(pro,2), fontsize=20)
plt.show()
```

# Ejemplos de variable Weibull

La variable aleatoria Weibull es una distribución de probabilidad continua que se utiliza frecuentemente en análisis de confiabilidad, estudios de vida útil de productos y en ingeniería para modelar el tiempo hasta el fallo de un componente o sistema

## 1. Ingeniería de Confiabilidad en Aeronaves


En la ingeniería aeronáutica, se podria utilizar la distribución Weibull para modelar el tiempo hasta el fallo de componentes críticos como motores y sistemas hidráulicos. Boeing, una de las principales empresas fabricantes de aviones del mundo, podria estar interesada en analizar la confiabilidad de los motores de su avión comercial estudiando la vida útil de un componente del motor de un avión. Usamos una distribución Weibull, Esto indica que la tasa de fallos aumenta con el tiempo, lo cual es típico para componentes que se desgastan debido a condiciones de operación severas.

## 2. Análisis de Durabilidad de Materiales


Empresas como Cemex, que se dedica a la elaboración de cementos, podria ocuparla en la ciencia de materiales, la distribución Weibull se usa para modelar la resistencia de materiales y predecir el tiempo hasta que estos se fracturen bajo tensión o estrés. Consideremos un nuevo tipo de material compuesto usado en la construcción de puentes. Se somete a pruebas de tensión para determinar su vida útil.

## 3. Estudios de Fiabilidad de Software


En ingeniería de software, la distribución Weibull se puede utilizar para modelar el tiempo entre fallos en sistemas de software, especialmente en fases de prueba y depuración. Oracle empresa dedicada a la creación y manejo de base de datos, estaria interesada en medir una estratefia de mantenimiento: Planificar las actualizaciones y mantenimientos para minimizar el tiempo de inactividad de su sistema ya que un tiempo alto de inactividad de su sistema podria derivar en perdidas de dinero y manejo de recursos de todos sus clientes.

## 4. Análisis de Supervivencia en Medicina


En estudios médicos, la distribución Weibull se emplea para modelar el tiempo hasta que ocurren eventos como la recaída de una enfermedad o la muerte de pacientes bajo tratamiento. Consideremos una empresa farmacéutica, que ha desarrollado un nuevo medicamento para tratar el cáncer de pulmón. La empresa está interesada en analizar el tiempo de supervivencia de los pacientes que reciben este tratamiento utilizando la distribución Weibull, basándose en datos de ensayos clínicos que han recopilado datos de supervivencia de pacientes que han participado en ensayos clínicos del nuevo medicamento para asi tener un mejor Desarrollo de Tratamientos, planificar una mejor Intervenciones Médicas si es que se requiere en un futuro, Evaluación de Riesgos en un futuro por posible recaida.

## 5. Modelado de Riesgo en Finanzas

En el ámbito financiero,un banco podria estar interesa en hacer modelos de riesgos de credito con la distribución Weibull, puede usarse para modelar el tiempo hasta la ocurrencia de eventos de riesgo, como el incumplimiento de pago de un préstamo o la quiebra de una empresa, buscando  modelar el tiempo hasta el incumplimiento de pago de una cartera de préstamos comerciales. Esto ayudará al banco a evaluar y gestionar los riesgos asociados con su cartera de préstamos. Pudiendo prevenir una buena Gestión de Riesgos, Provisión para Pérdidas, una Evaluación de la Solvencia mas excata, y quiza desarrollar una mejor Estrategias de Cobranza.
